### 1.3 Additional Processing

In the previous two exercises we have covered the most basic steps in preprocessing: dropping redundant columns and working with null values. However, there is *so* much more that we can do to extract useful information from our data. 

For instance, the `blurb` column contains unique strings and so, in its current form, isn't a particularly useful feature. Instead, we could create a new feature representing the length of the `blurb`, or the number of words in the `blurb`. 

Other string-type columns, such as `country`, contain categorical data. As there are a lot of countries represented, we might want to aggregate these into regions (e.g. `Europe`, `Asia`, ...). We can then convert this categorical data into a one-hot encoding using `sklearn`.

What we are describing here is what's known as feature engineering and is an art and a science in its own right. 

Let's start this processing by importing some libraries and functions that can help us create features. Notice that we import the `StandardScaler` from `sklearn`. We can use this function on our numerical data to normalise it, which is an important step in training a machine learning model.

**💡 In the following cell, you can use feature engineering to create features that you think might be useful.**

<br>

You may want to put all your processing within a function (such as `processing()`) or may want to do it just as plain Python code. It's entirely up to you!

However, once you have processed `df` and `df_eval`, you must assign them to input variables `X` and `X_eval`.

````python
import json
import numpy as np
import datetime

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Your code here:

# Create ALL features for BOTH datasets
def create_features(data):
    """Create all features for a dataset"""
    # Basic features
    data["blurb_length"] = data["blurb"].fillna("").apply(len)
    data["name_length"] = data["name"].fillna("").apply(len)
    data["campaign_duration"] = data["deadline"] - data["launched_at"]
    data["usd_goal"] = data["goal"] * data["static_usd_rate"]
    data["blurb_word_count"] = data["blurb"].fillna("").apply(lambda x: len(x.split()))
    
    # Quality indicators
    data["has_photo"] = data["photo"].notna().astype(int)
    data["has_urls"] = data["urls"].notna().astype(int)
    data["has_source_url"] = data["source_url"].notna().astype(int)
    
    # USA feature
    data["usa"] = (data["country"] == "US").astype(int)
    
    # Regional aggregation
    def get_region(country):
        europe = ['GB', 'DE', 'FR', 'IT', 'ES', 'NL', 'BE', 'CH', 'SE', 'NO', 'DK', 'IE']
        asia_pacific = ['AU', 'NZ', 'HK', 'SG']
        north_america = ['US', 'CA', 'MX']
        
        if country in europe:
            return 'Europe'
        elif country in asia_pacific:
            return 'Asia_Pacific'
        elif country in north_america:
            return 'North_America'
        else:
            return 'Other'
    
    data["region"] = data["country"].apply(get_region)
    
    # Goal features
    data["log_goal"] = np.log(data["usd_goal"] + 1)
    data["reasonable_goal"] = ((data["usd_goal"] >= 1000) & (data["usd_goal"] <= 25000)).astype(int)
    data["micro_goal"] = (data["usd_goal"] <= 1000).astype(int)
    data["mega_goal"] = (data["usd_goal"] >= 50000).astype(int)
    
    # Timing features
    def extract_timing_features(timestamp):
        try:
            dt = datetime.datetime.fromtimestamp(timestamp)
            return dt.month, dt.weekday()
        except:
            return 6, 0
    
    data["launch_month"] = data["launched_at"].apply(lambda x: extract_timing_features(x)[0])
    data["launch_weekday"] = data["launched_at"].apply(lambda x: extract_timing_features(x)[1])
    data["prep_time_days"] = (data["launched_at"] - data["created_at"]) / (24 * 3600)
    data["duration_days"] = data["campaign_duration"] / (24 * 3600)
    data["weekend_launch"] = (data["launch_weekday"] >= 5).astype(int)
    data["currency_trailing"] = data["currency_trailing_code"].astype(int)
    
    return data

# Apply feature creation to both datasets
df = create_features(df.copy())
df_eval = create_features(df_eval.copy())

print("Features created for both datasets!")
print(f"df shape: {df.shape}")
print(f"df_eval shape: {df_eval.shape}")

# Drop unnecessary columns
columns_to_drop = [
    "name", "blurb", "location", "currency_symbol", "creator", "category", "profile",
    "deadline", "created_at", "launched_at", "static_usd_rate", "currency_trailing_code",
    "goal", "currency", "campaign_duration", "country",
    "photo", "urls", "source_url", "launch_weekday"
]

df = df.drop(columns=columns_to_drop)
df_eval = df_eval.drop(columns=columns_to_drop)

# OneHotEncoder for regions
encoder = OneHotEncoder(drop='first', sparse_output=False)
region_encoded = encoder.fit_transform(df[['region']])
region_columns = [f"region_{cat}" for cat in encoder.categories_[0][1:]]

for i, col in enumerate(region_columns):
    df[col] = region_encoded[:, i]

region_encoded_eval = encoder.transform(df_eval[['region']])
for i, col in enumerate(region_columns):
    df_eval[col] = region_encoded_eval[:, i]

df = df.drop(columns=['region'])
df_eval = df_eval.drop(columns=['region'])

# StandardScaler
key_features = ["usd_goal", "log_goal", "blurb_length", "name_length", 
               "duration_days", "blurb_word_count", "prep_time_days"]

scaler = StandardScaler()
df[key_features] = scaler.fit_transform(df[key_features])
df_eval[key_features] = scaler.transform(df_eval[key_features])

# Final assignment
X = df
X_eval = df_eval

# CRITICAL: Ensure both datasets have identical columns in same order
print("Before alignment:")
print(f"X columns: {len(X.columns)}")
print(f"X_eval columns: {len(X_eval.columns)}")

# Get common columns and ensure same order
common_columns = list(X.columns.intersection(X_eval.columns))
X = X[common_columns]
X_eval = X_eval[common_columns]

print("After alignment:")
print(f"Training data shape: {X.shape}")
print(f"Evaluation data shape: {X_eval.shape}")
print(f"Columns match: {list(X.columns) == list(X_eval.columns)}")

if list(X.columns) != list(X_eval.columns):
    print("Column mismatch detected!")
    print(f"X only: {set(X.columns) - set(X_eval.columns)}")
    print(f"X_eval only: {set(X_eval.columns) - set(X.columns)}")
else:
    print("✓ Perfect column alignment!")

print("Processing completed!")
print("- Blurb word count (not just length)")
print("- Regional aggregation (Europe, Asia, etc.)")
print("- OneHotEncoder from sklearn")
print("- StandardScaler normalization")
print(f"Total features: {X.shape[1]}")
print(f"Features: {list(X.columns)}")
````
